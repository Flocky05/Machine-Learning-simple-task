```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Google colab projects/carprices.csv')
```

```python
df
```

```python
dummies=pd.get_dummies(df['Car Model'])
```

```python
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
```

```python
merged
```

```python
final=merged.drop(['Car Model'],axis='columns')
```

```python
final
```

```python
y=df['Sell Price']
```

```python
y
```

```python
x=final.drop(['Sell Price'],axis='columns')
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[45000,5,1,0,0,0]])
```
